```python
import pandas as pd
# Load the data
df = pd.read_csv('./sample_data/dow_jones_index.csv')
# Display the first 10 rows
df.head(n=10)

```

```python
df.columns.to_list()
```

```python
lines, columns = df.shape
print(f'Number of lines: {lines}')
print(f'Number of columns: {columns}')
```

```python
# function to filter and select data
def filter_and_select_stock_data(df, stock_symbol):
    filtered_df = df[df['stock'] == stock_symbol]
    selected_columns_df = filtered_df[['date', 'open', 'high', 'low', 'close']]
    return selected_columns_df
```

```python
# using the function for 'MCD' and 'KO'
df_mcd = filter_and_select_stock_data(df, 'MCD')
df_ko = filter_and_select_stock_data(df, 'KO')

# Display the first 10 rows for each
print(df_mcd.head(n=10))
print(df_ko.head(n=10))
```

```python
# Function to conversion a string to float after splitting by '$'
def try_convert(value):
    try:
        if isinstance(value, str):
            return float(value.split(sep='$')[-1])
        return value
    except ValueError:
        # Log the error if conversion fails
        print(f"Conversion failed for value: {value}")
        return None

# Convert string values to float in specified columns
for col in ['open', 'high', 'low', 'close']:
    df_mcd[col] = df_mcd[col].apply(lambda value: try_convert(value))
    df_ko[col] = df_ko[col].apply(lambda value: try_convert(value))
    
# Print DataFrame and data types after all conversions
print(df_mcd)
print(df_ko)
print(df_mcd.dtypes)
print(df_ko.dtypes)
```

```python
import seaborn as sns
```

```python
plot = sns.lineplot(x='date', y='open', data=df_mcd)
plot.set_xticklabels(labels=df_mcd['date'], rotation=90)
```

```python
plot = sns.lineplot(x='date', y='open', data=df_ko)
plot.set_xticklabels(labels=df_ko['date'], rotation=90)
```

```python
# here is the plot(graphical representation of data designed) for mcd stock
plot = sns.lineplot(x='date', y='value', hue='variable', data=pd.melt(df_mcd, ['date']))
# Rotate the x-axis labels for better readability
plot.set_xticklabels(labels=df_mcd['date'], rotation=90)
# Save the plot to a png file
plot.figure.savefig('./sample_data/mcd.png')
```

From the data of the MCD stock, we can extract several insights:

1. **Opening and Closing Prices**: We can analyze the opening and closing prices of the MCD stock over time to identify any trends or patterns. This can help us understand the stock's performance and potential investment opportunities.

2. **High and Low Prices**: By examining the high and low prices of the MCD stock, we can determine the stock's volatility and the range within which it traded during a specific period. This information can be useful for setting stop-loss or take-profit levels.

3. **Date**: The date column allows us to track the stock's performance over time and identify any seasonal or cyclical patterns that may exist.

4. **Visualization**: On the graphical representation of MCD data, we can visualize a slight increase in the stock's price during the evaluated timeframe
